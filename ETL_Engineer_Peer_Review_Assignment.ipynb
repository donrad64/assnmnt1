## Transform

Using <code>exchange_rate</code> and the `exchange_rates.csv` file find the exchange rate of USD to GBP. Write a transform function that

1. Changes the `Market Cap (US$ Billion)` column from USD to GBP
2. Rounds the Market Cap (US$ Billion)` column to 3 decimal places
3. Rename `Market Cap (US$ Billion)` to `Market Cap (GBP$ Billion)`


```python

def transform(exchange_rate):
    # Load the exchange_rates.csv file into a DataFrame
    df = extract()
    
    # Convert the Market Cap column from USD to GBP
    df['Market Cap (US$ Billion)'] = pd.to_numeric(df['Market Cap (US$ Billion)'], errors='coerce')*exchange_rate
    
    # Round the Market Cap column to 3 decimal places
    df['Market Cap (US$ Billion)'] = df['Market Cap (US$ Billion)'].round(3)
    
    # Rename the Market Cap column to Market Cap (GBP$ Billion)
    df.rename(columns={'Market Cap (US$ Billion)': 'Market Cap (GBP$ Billion)'}, inplace=True)
    
    return df



# Call the transform function
transformed_df = transform(exchange_rate)

# Print the transformed DataFrame
print(transformed_df)

    
```

## Logging Function


Write the logging function <code>log</code> to log your data:


```python
import logging

def log(data):
    # Create a logger
    logger = logging.getLogger('data_logger')
    logger.setLevel(logging.INFO)
    
    # Define the log file name and format
    log_file = 'data.log'
    formatter = logging.Formatter('%(asctime)s - %(levelname)s - %(message)s')
    
    # Create a file handler and set the formatter
    file_handler = logging.FileHandler(log_file)
    file_handler.setFormatter(formatter)
    
    # Add the file handler to the logger
    logger.addHandler(file_handler)
    
    # Log the data
    logger.info(data)
    
    # Close the file handler
    file_handler.close()
    
    # Remove the file handler from the logger
    logger.removeHandler(file_handler)
    
    print("Data logged successfully.")

# Example usage:
data = "Sample data to log"
log(data)
```
